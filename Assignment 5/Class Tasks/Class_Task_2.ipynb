```python
class Farmer:
  def __init__(self, name_id = None):
    self.name_id = name_id
    if self.name_id == None:
      print("Welcome to your farm!")
    elif type(self.name_id) == str:
      print(f"Welcome to your farm, {self.name_id}!")
    elif type(self.name_id) == int:
      print(f"Welcome to your farm. Your farm ID is {self.name_id}!")
    self.crops = []
    self.fishes = []
  def addCrops(self, *crop):
    self.crops.extend(crop)
    if len(crop) == 0:
      print("No crop(s) added.")
    else:
      print(f"{len(crop)} crop(s) added.")
  def addFishes(self, *fish):
    self.fishes.extend(fish)
    if len(fish) == 0:
      print("No fish added.")
    else:
      print(f"{len(fish)} fish(s) added.")
  def showGoods(self):
    crops = ""
    fishes = ""
    if len(self.crops) == 0:
      print("You don't have any crop(s).")
    else:
      print(f"You have {len(self.crops)} crop(s):")
      for item in range(len(self.crops)):
        if item == len(self.crops) - 1:
          crops += self.crops[item]
        else:
          crops += self.crops[item] + ", "
      print(crops)
    print(f"You have {len(self.fishes)} fish(s):")
    if len(self.fishes) == 0:
      print("You don't have any fish(s).")
    else:
      for item in range(len(self.fishes)):
        if item == len(self.fishes) - 1:
          fishes += self.fishes[item]
        else:
          fishes += self.fishes[item] + ", "
      print(fishes)

f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```
