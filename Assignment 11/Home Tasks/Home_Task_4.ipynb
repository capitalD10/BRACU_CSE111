```python
class Cakes:
  order_list = {}
  reviews = {}
  def __init__(self, flavor, weight, sweet = "Moderate", cream = "Whipped Cream"):
    self.flavor = flavor + " Cake"
    self.weight = weight
    self.sweet = sweet
    self.cream = cream
    self.price = (weight * 1200) / 1000
    if self.flavor not in Cakes.order_list:
      Cakes.order_list[f"{self.flavor} {self.weight}gm"] = 1
    else:
      Cakes.order_list[f"{self.flavor} {self.weight}gm"] += 1
  def add_customization(self, sugar, cream):
    self.sweet = sugar
    self.cream = cream
  def cake_details(self):
    print(f"Flavor: {self.flavor}, Weight: {self.weight} gm")
    print(f"Sweetness: {self.sweet} sugar, Cream Type: {self.cream}")
    print(f"Price: {self.price} Taka")
  @classmethod
  def give_feedbacks(cls, cake, review):
    if cake not in cls.reviews:
      cls.reviews[cake] = [review]
    else:
      cls.reviews[cake].append(review)
    print("Thanks for your valuable feedback!")
  @classmethod
  def show_feedbacks(cls):
    print(cls.reviews)
class Cheese_Cakes(Cakes):
  def __init__(self, cake, weight, bake = "baked"):
    cake += " Cheese"
    super().__init__(cake, weight, "Moderate", "Cream Cheese")
    self.c_type = bake
    self.price = (weight * 1500) / 1000
  def add_customization(self):
    print("Sorry! No customization available for cheese cakes.")
  def cake_details(self):
    print(f"Flavor: {self.flavor}, Weight: {self.weight} gm")
    print(f"Sweetness: {self.sweet} sugar, Cream Type: {self.cream}")
    print(f"Cake Type: {self.c_type}, Price: {self.price} Taka")
  @classmethod
  def give_feedbacks(self, cake, review):
    super().give_feedbacks(cake, review)
    print("You will get 10% discount for your next purchase!")

order_1=Cakes("Chocolate",500)
order_2=Cakes("Vanilla",800)
print("(1)**********************************")
order_1.cake_details()
print("(1.1)**********************************")
print(Cakes.order_list)
print("(2)**********************************")
order_2.add_customization("Zero","Butter Cream")
order_2.cake_details()
print("(3)**********************************")
Cakes.give_feedbacks("Chocolate Cake","Very Delicious")
Cakes.give_feedbacks("Chocolate Cake","Yummy")
print("(4)**********************************")
Cakes.show_feedbacks()
print("(5)**********************************")
ch_order1=Cheese_Cakes("Red velvet",700)
ch_order1.cake_details()
print("(6)**********************************")
ch_order1.add_customization()
print("(7)**********************************")
ch_order2=Cheese_Cakes("Blue Berry",900,"No Bake")
ch_order2.cake_details()
print("(8)**********************************")
print(Cakes.order_list)
print("(9)**********************************")
Cheese_Cakes.give_feedbacks("Red velvet Cheese Cake","Average")
Cheese_Cakes.give_feedbacks("Blue Berry Cheese Cake","Liked it")
print("(10)**********************************")
Cakes.show_feedbacks()
```
