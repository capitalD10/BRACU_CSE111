```python
class Student:
  def __init__(self, name, id, cgpa):
    self.name = name
    self.id = id
    self.cgpa = cgpa
    # self.students = []
    # self.students.append(self.id)
  def setId(self, new_id):
    self.id = new_id
class Department:
  def __init__(self, dept):
    self.dept = dept
    self.stud = 0
    self.everyone = {}
  def findStudent(self, stud_id):
    student_found = False
    for student in self.everyone:
      if self.everyone[student][0] == stud_id:
        print("Student info:")
        print(f"Student Name: {student}")
        print(f"ID: {self.everyone[student][0]}")
        print(f"CGPA: {self.everyone[student][1]}")
        student_found = True
        break
    if not student_found:
      print("Student with this ID doesn't exist, Please give a valid ID")
  def addStudent(self, *people):
    for person in people:
      student_exists = False
      for student in self.everyone:
        if  self.everyone[student][0] == person.id:
          student_exists = True
          break
      if not student_exists:
        self.everyone[person.name] = [person.id, person.cgpa]
        print(f"Welcome to {self.dept} department, {person.name}")
        self.stud += 1
      else:
        print("Student with the same ID already exists, Please try with another ID")
  def details(self):
    print(f"Department Name: {self.dept}")
    print(f"Number of student: {self.stud}")
    for every in self.everyone:
      print(f"Student name: {every}, ID: {self.everyone[every][0]}, cgpa: {self.everyone[every][1]}")

s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```
