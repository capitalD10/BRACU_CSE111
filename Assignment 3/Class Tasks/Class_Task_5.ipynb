```python
class CellPackage:
  def __init__(self, price, data, talk_time, messages, cashback, validity):
    self.price = price
    self.talk_time = talk_time
    self.messages = messages
    self.validity = validity
    datapack = ""
    for x in data:
      if x == " ":
        break
      else:
        datapack += x
    datapack = int(datapack)
    self.data = datapack * 1024
    cash = ""
    for x in cashback:
      if x == "%":
        break
      else:
        cash += x
    cash = int(cash)
    self.cashback = (price * cash) / 100




pkg = CellPackage(150, '6 GB', 99, 20, '7%', 7)
print('===========Package 1=============')
if pkg.data == 0:
  pass
else:
  print(f"Data = {pkg.data} MB")
if pkg.talk_time == 0:
  pass
else:
  print(f"Talktime = {pkg.talk_time} minutes")
if pkg.messages == 0:
  pass
else:
  print("SMS/MMS =", pkg.messages)
print(f"Validity = {pkg.validity} days")
print(f"--> Price = {pkg.price} tk")
if pkg.cashback == 0:
  pass
else:
  print(f"Buy now to get {round(pkg.cashback)} tk cashback.")

pkg2 = CellPackage(700, '35 GB', 700, 0, '10%', 30)
print('===========Package 2=============')
if pkg2.data == 0:
  pass
else:
  print(f"Data = {pkg2.data} MB")
if pkg2.talk_time == 0:
  pass
else:
  print(f"Talktime = {pkg2.talk_time} minutes")
if pkg2.messages == 0:
  pass
else:
  print("SMS/MMS =", pkg2.messages)
print(f"Validity = {pkg2.validity} days")
print(f"--> Price = {pkg2.price} tk")
if pkg2.cashback == 0:
  pass
else:
  print(f"Buy now to get {round(pkg2.cashback)} tk cashback.")

pkg4 = CellPackage(120, '0 GB', 190, 0, '0%', 10)
print('============Package 3============')
if pkg4.data == 0:
  pass
else:
  print(f"Data = {pkg4.data} MB")
if pkg4.talk_time == 0:
  pass
else:
  print(f"Talktime = {pkg4.talk_time} minutes")
if pkg4.messages == 0:
  pass
else:
  print("SMS/MMS =", pkg4.messages)
print(f"Validity = {pkg4.validity} days")
print(f"--> Price = {pkg4.price} tk")
if pkg4.cashback == 0:
  pass
else:
  print(f"Buy now to get {round(pkg4.cashback)} tk cashback.")
```
